### Aggregation

```python
# aggregation  -> one class owns a class -> has a relation -> like restaurant has a menu

# example
class Customer:

  def __init__(self,name,gender,address):
    self.name = name
    self.gender = gender
    self.address = address

  def print_address(self):
    # can't access private attributes
    # print(self.address.__city,self.address.pin,self.address.state)
    # for accessing private attributes
    # print(self.address.get_city(),self.address.pin,self.address.state)
    # 2nd method to access private attributes -> this is not good practice
    # print(self.address._Address__city,self.address.pin,self.address.state)
    return (self.address.city,self.address.pin,self.address.state)

  def edit_profile(self,new_name,new_city,new_pin,new_state):
    self.name = new_name
    self.address.edit_address(new_city,new_pin,new_state)

class Address:

  def __init__(self,city,pin,state):
    self.city = city
    # can't access private attributes
    # self.__city = city
    self.pin = pin
    self.state = state

  # for accessing private attributes -> make get method
  def get_city(self):
    return self.__city

  # for accessing methods
  def edit_address(self,new_city,new_pin,new_state):
    self.city = new_city
    self.pin = new_pin
    self.state = new_state

add = Address('New Delhi','110044','Delhi')
cust = Customer('Ravi','Male',add)

cust.print_address()

cust.edit_profile('Amit','Jaipur','214530','Rajasthan')

cust.print_address()
```

### Inheritance

```python
# aggregation class diagram link by a diamond sign
# inheritance class diagram link by a triangle sign
```

```python
# inheritance -> two classes as parent and child -> child class can access parent class
# parent class
class User:

  def __init__(self):
    self.name = "Ravi"

  def login(self):
    print("Login")

# child class
class Student(User):
  # if we provide constructor in child class then we can't able to access parent constructor
  # using super method we can access parent constructor aslo
  def enroll(self):
    print('enroll in the class.')

u = User()
s = Student()

print(s.name)
s.login()
s.enroll()
```

```python
# what get inherited?
# 1. constructor
# 2. non private attributes
# 3. non private methods
```

```python
# constructor example 1
# if there is no constructor in child class then it will access parent class constructor

class Phone:

  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera
    print('phone constructor')

  def buy(self):
    print('buy')

class SmartPhone(Phone):
  pass

s = SmartPhone(22000,'Nothing',50)
s.buy()
print(s.price)
```

```python
# constructor example 2
# if there is constructor in child class then it can't access the parent class constructor and never initialize the value

class Phone:

  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera
    print('phone constructor')

class SmartPhone(Phone):

  def __init__(self,os,ram):
    self.os = os
    self.ram = ram
    print('smartphone constructor')

s = SmartPhone('Android',8)
```

```python
# child class can't access private attributes and methods

class Phone:

  def __init__(self,price,brand,camera):
    self.__price = price
    self.brand = brand
    self.camera = camera
    print('phone constructor')

  # def __show(self):
    # return (self.__price)

  def show(self):
    return (self.__price)

class SmartPhone(Phone):

  # this will show an error bcos we can't access private attributes
  # def get(self):
    # return (self.__price)
  pass

s = SmartPhone(22000,'Nothing',50)
s.show()
```

```python
# method overriding -> there are same method in child and parent class -> this will execute child class

class Phone:

  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print('buying a phone')

class SmartPhone(Phone):

  def buy(self):
    print('buying a smartphone')

s = SmartPhone(22000,'Nothing',50)
s.buy()
```

### Super Keyword

```python
# super keyword -> to access both classes

class Phone:

  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print('buying a phone')

class SmartPhone(Phone):

  def buy(self):
    print('buying a smartphone')
    super().buy()

s = SmartPhone(22000,'Nothing',50)
s.buy()
```

```python
# example super keyword

class Phone:

  def __init__(self,price,brand,camera):
    print('inside phone constructor')
    self.price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print('buying a phone')

class SmartPhone(Phone):

  def __init__(self,price,brand,camera,os,ram):
    print('inside smartphone constructor')
    super().__init__(price,brand,camera)
    self.os = os
    self.ram = ram
    print('inside smartphone constructor')

s = SmartPhone(22000,'Nothing',50,'Android',8)
print(s.os)
print(s.brand)
```
