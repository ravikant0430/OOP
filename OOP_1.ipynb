# Class

```python
# making own class
from types import new_class
import getpass

class Atm:

  # constructor
  def __init__(self):
    self.pin = ""
    self.balance = 0

  def menu(self):
   user_input = int(input("""
   1. Press 1 to Create Pin
   2. Press 2 to Change Pin
   3. Press 3 to Check Balance
   4. Press 4 to Withdraw
   5. Press 5 to Deposit
   6. Press Anything to Exit
   """))

   if user_input == 1:
    # create pin
      self.create_pin()
   elif user_input == 2:
    # change pin
      self.change_pin()
   elif user_input == 3:
    # check balance
      self.check_balance()
   elif user_input == 4:
    # withdraw
      self.withdraw()
   elif user_input == 5:
    # deposit
      self.deposit()
   else:
      exit()

  def create_pin(self):
   user_pin = getpass.getpass("Create New Pin")
   self.pin = user_pin
   print("Pin Created Successfully.")
   balance = int(input("Enter Balance"))
   self.balance = balance
   self.menu()

  def change_pin(self):
    old_pin = getpass.getpass("Enter Old Pin")
    if old_pin == self.pin:
      new_pin = getpass.getpass("Enter New Pin")
      self.pin = new_pin
      print("Pin Changed Successfully.")
    else:
      print("Incorrect Pin")
    self.menu()

  def check_balance(self):
    user_pin = getpass.getpass("Enter Pin")
    if user_pin == self.pin:
      print(f"Your Balance is {self.balance}")
    else :
      print("Incorrect Pin")
    self.menu()

  def withdraw(self):
    user_pin = getpass.getpass("Enter Pin")
    if user_pin == self.pin:
      withdraw = int(input("Enter Balance"))
      if withdraw<self.balance:
        self.balance = self.balance - withdraw
        print("Withdraw Successful")
      else:
        print("Insufficient Balance")
    else:
      print("Incorrect Pin")
    self.menu()


  def deposit(self):
    user_pin = getpass.getpass("Enter Pin")
    if user_pin == self.pin:
      deposit = int(input("Enter Balance"))
      self.balance = self.balance + deposit
      print("Deposit Successful")
    else:
      print("Incorrect Pin")
    self.menu()



```

# making class for rectangle solution

```python
class rectangle:
   def __init__(self,x,y):
    self.x = x
    self.y = y

   def area(self):
    return self.x*self.y

   def parameter(self):
    return 2*(self.x+self.y)
```

```python
rect = rectangle(2,3)
```

```python
rect.area()
```

```python
rect.parameter()
```

# making class to solve fraction problem

```python
class fraction:

 # parameterized constructor -> this takes parameter
  def __init__(self,n,d):
    self.n = n
    self.d = d

  def __str__(self):
    return "{}/{}".format(self.n,self.d)

  def __add__(self,other):
    return "{}/{}".format(self.n*other.d+self.d*other.n,self.d*other.d)

  def __sub__(self,other):
    return "{}/{}".format(self.n*other.d-self.d*other.n,self.d*other.d)

  def __mul__(self,other):
    return "{}/{}".format(self.n*other.n,self.d*other.d)

  def __truediv__(self,other):
    return "{}/{}".format(self.n*other.d,self.d*other.n)
```

```python
fr = fraction(1,2)
fr1 = fraction(1,2)
```

```python
print(fr+fr1)
```

```python
print(fr-fr1)
```

```python
print(fr*fr1)
```

```python
print(fr/fr1)
```
